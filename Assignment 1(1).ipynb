# box plot

A box plot is a graphical representation used to show the distribution of a dataset. It displays the summary of a dataset based on five key summary statistics: the minimum, first quartile (Q1), median (Q2), third quartile (Q3), and maximum. It also helps to identify outliers and the spread of the data.

```python
plt.figure(figsize=(10, 9))  
plt.boxplot(df['Discount Amount'])
plt.title('Box plot')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

 Here all the features will be transfromed in such a way that it will have the properties of a standard normal distribution with mean= 0 and standard deviation = 1.
 
Formula for Z-Score Normalization:
        z = x - mean/  standard deviation

Standardization scales the data to have a mean of 0 and standard deviation of 1.

# conversion of categorical data into Dummy variable


Encoding refers to the process of converting categorical data (non-numerical data) into numerical form so that machine learning models can interpret and use it effectively. Many machine learning algorithms require numerical input, so encoding is essential when working with categorical features in your dataset.


One-Hot Encoding:

One-hot encoding creates binary (0 or 1) columns for each unique category in the feature. Each category gets its own column with a value of 1 where the category is present and 0 where it is not.

Best suited for: Nominal data (categories without an inherent order), like "City" or "Product Type

USING ON HOT ENCODING TYPE

STPE 1: CREATE DUMMY VARIABLE COLUMNS & store variable like dummies


```python
dummies =pd.get_dummies(df[['Brand','SKU','BU','Model']])
dummies
```

STPE :2 CONCATENATE THIS DUMMY'S DATA FROM WITH MY ORIGINAL DATAFRAME USING MERGED

```python
merged = pd.concat([df, dummies], axis= 'columns')
merged
```

# Conclusion

1. Descriptive Analytics and Visualizations:

2. Importance of Data Preprocessing:

One-Hot Encoding is used to convert categorical data into numerical form so that models can understand them. This is crucial for treating categories as separate entities rather than giving them a numerical order.

 Data preprocessing steps like standardization and one-hot encoding are important because they prepare the data for analysis and ensure the machine learning models work effectively with the data.
